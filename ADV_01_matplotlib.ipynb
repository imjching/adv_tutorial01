Basic Plot
----------------
* First let try to <b>import</b> matplitlib package

```python
import matplotlib.pyplot as plt
```

* Let's define some data
* The following data represent total number of reported death by Dengue in Malaysia. [link](http://www.data.gov.my/data/en_US/dataset/jumlah-kes-kematian-akibat-wabak-denggi-yang-telah-dilaporkan-di-malaysia)

```python
x = [0, 1, 2, 3, 4, 5]
y = [109, 33, 35, 85, 179, 110]
```

* Now let's plot it

```python
plt.plot(x, y)
```

```python
plt.show()
```

What just happend? The default behavior will show the graph in a new windows!

* Now let's try to make the graph appear inline

```python
%matplotlib inline
```

* The line above configures matplotlib to show figures embedded in the notebook, instead of opening new windows.
* Now let's try to show the graph again

```python
plt.plot(x, y)
plt.show()
```

* Let's change the line style
* (refer to documentation for more options: http://matplotlib.org/api/index.html)

```python
plt.plot(x, y, linestyle="dashed", marker="o", color="green")
```

* Let's add a title and labels to our graph

```python
plt.plot(x, y, linestyle="dashed", marker="o", color="green")
plt.xlabel('Year')
plt.ylabel('Cases')
plt.title('Number of death by Dengue in Malaysia')
plt.show()
```

* Configure X and Y axis

```python
# set the limit of X axis
plt.xlim(-1,8)
# set the ticks of X axis
plt.xticks([1, 2, 3, 4])

# set the limit of Y axis
plt.ylim(0, 200)
plt.yticks([0, 100, 200])
plt.plot(x, y, linestyle="dashed", marker="o", color="green")
```

* Setting the ticket labels

```python
x_labels = ['2010', '2011', '2012', '2013', '2014', '2015']
plt.xticks([0, 1, 2, 3, 4, 5], x_labels, rotation=45)  # options for rotation: angle in degrees | vertical | horizontal
plt.plot(x, y, linestyle="dashed", marker="o", color="green", label="death cases")
```

* Plotting 2 lines with legends
* Using another set of data: total reported dengue cases. [Link](http://www.data.gov.my/data/en_US/dataset/jumlah-kes-keseluruhan-wabak-denggi-yang-telah-dilaporkan-di-malaysia)

```python
total_cases = [38943, 20037, 21900, 40831, 87503, 36356]
plt.plot(x, y, linestyle="dashed", marker="o", color="green", label="death cases")
plt.plot(x, total_cases, linestyle="dashed", marker="o", color="red", label="total cases")
plt.xticks([0, 1, 2, 3, 4, 5], x_labels, rotation=45)
plt.legend(loc=2);       #
```

* We can use other method to solve the scaling problem. 
* For example, we can add new axes

```python
fig = plt.figure()

# Params = left, bottom, width, height
axes1 = fig.add_axes([0, 0, 1, 1]) 
axes2 = fig.add_axes([0, 1.4, 1, 1]) # inset axes

axes1.plot(x, total_cases, 'r', marker='x')
axes1.set_xlabel('Year')
axes1.set_ylabel('cases')
axes1.set_title('Total dengue cases')
axes1.set_xticklabels(x_labels, rotation=45)
axes1.grid(True)

axes2.plot(x, y, 'g', marker='o')
axes2.set_xlabel('Year')
axes2.set_ylabel('cases')
axes2.set_title('Total death');
axes2.set_xticklabels(x_labels, rotation=45)
```

* Changing configuration

```python
import matplotlib
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'serif'})
plt.plot(x, y, linestyle="dashed", marker="o", color="green", label="death cases")
```

* Saving figures to an image

```python
# reduce font size
matplotlib.rcParams.update({'font.size': 10, 'font.family': 'serif'})

plt.xticks([0, 1, 2, 3, 4, 5], x_labels, rotation=45)
plt.plot(x, y, linestyle="dashed", marker="o", color="g")

plt.savefig('dengue_death.png', transparent=True, bbox_inches='tight', pad_inches=0)
```

* Save multiple figures (only support PDF)

```python
# method 1
from matplotlib.backends.backend_pdf import PdfPages
pp = PdfPages('save_figure.pdf')

matplotlib.rcParams.update({'font.size': 12, 'font.family': 'serif'})
fig = plt.figure()
axes1 = fig.add_axes([0.2, 0.2, 0.4, 0.4])

axes1.plot(x, total_cases, 'r', marker='x')
axes1.set_xlabel('Year')
axes1.set_ylabel('cases')
axes1.set_title('Total dengue cases')
axes1.set_xticklabels(x_labels, rotation=45)
axes1.grid(True)
pp.savefig(fig)

plt.close()
fig = plt.figure()
axes1 = fig.add_axes([0.2, 0.2, 0.4, 0.4])

axes1.plot(x, y, 'r', marker='x')
axes1.set_xlabel('Year')
axes1.set_ylabel('cases')
axes1.set_title('Total Death')
axes1.set_xticklabels(x_labels, rotation=45)
axes1.grid(True)
pp.savefig(fig)

pp.close()

    
```

```python
# method 2
with PdfPages('save_plot.pdf') as pdf:
    plt.xticks([0, 1, 2, 3, 4, 5], x_labels, rotation=45)
    plt.title = "Page 1"
    plt.xlabel = "Year"
    plt.ylabel = "Cases"
    plt.ylim(0, 200)
    plt.plot(x, y, linestyle="dashed", marker="o", color="g")
    plt.savefig(pdf, format='pdf')
#     pdf.savefig(plt.figure())
    plt.close()
    
    plt.xticks([0, 1, 2, 3, 4, 5], x_labels, rotation=45)
    plt.ylim(0, 100000)
    plt.title = "Page 1"
    plt.xlabel = "Year"
    plt.ylabel = "Cases"
    plt.plot(x, total_cases, linestyle="dashed", marker="o", color="g")
    plt.savefig(pdf, format='pdf')
    plt.close()
```
